```python
import pandas as pd
from pandas import datetime
import matplotlib.pyplot as plt
```

```python
series=pd.read_csv("E:\\notebooks file\\sales.csv",header=0,index_col=0,parse_dates=True,squeeze=True)
```

```python
series
```

```python
upsampled=series.resample('D').mean()
print(upsampled.head(32))
```

```python
interpolated=upsampled.interpolate(method='linear')
print(interpolated.head(32))
interpolated.plot()
plt.show()
```

```python
resample=series.resample('Q')
quartely_mean_sales=resample.mean()
```

```python
print(quartely_mean_sales.head())
quartely_mean_sales.plot()
plt.show()
```

```python
series=pd.read_csv("E:\\notebooks file\\airline-passengers.csv",header=0,index_col=0,parse_dates=True,squeeze=True)
```

```python
plt.subplot(211)
plt.plot(series)
plt.subplot(212)
plt.hist(series)
```

```python
from pandas import DataFrame
from numpy import sqrt
```

```python
dataframe=DataFrame(series.values)
dataframe.columns=['passengers']
dataframe['passengers']=sqrt(dataframe['passengers'])
```

```python
plt.subplot(211)
plt.plot(dataframe['passengers'])
plt.subplot(212)
plt.hist(dataframe['passengers'])
```

```python
from numpy import log
dataframe=DataFrame(series.values)
dataframe.columns=['passengers']
dataframe['passengers']=log(dataframe['passengers'])
```

```python
plt.subplot(211)
plt.plot(dataframe['passengers'])
plt.subplot(212)
plt.hist(dataframe['passengers'])
```
